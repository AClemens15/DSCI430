Part 7
- make a list of common column names

```python
all_items = columnssets[0]
for columns in columnssets[1:]:
    all_items = all_items.union(columns)
all_items
```

```python
common_items = all_items
for columns in columnssets[1:]:
    common_items = common_items.intersection(columns)
common_items
```

Part 8 
- determine what files have extra column names

```python
totals = [(year, columns.difference(common_items)) for year, columns in zip(range(2002,2016), columnssets)]
totals
```
